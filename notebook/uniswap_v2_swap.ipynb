```python
def calc_swap(reserve_in, reserve_out, amount_in, f):
    return amount_in * (1 - f) * reserve_out / (reserve_in + amount_in * (1 - f))

# x = ETH, y = DAI
x = 1000
y = 2000000
f = 0.003

spot0 = y / x
print("x", x)
print("y", y)
print("spot", spot0)
print("-------------")

# Swap DAI (y) for ETH (x)
dy0 = 10 * 1e6
dy = dy0
dx = calc_swap(y, x, dy, f)
x -= dx
y += dy
spot1 = y / x

print("swap DAI -> ETH")
print("dx", -dx)
print("dy", dy)
print("x", x)
print("y", y)
print("spot1", spot1)
print("-------------")

# Swap ETH (x) for DAI (y)
dy = calc_swap(x, y, dx, f)
dy1 = dy
x += dx
y -= dy
spot2 = y / x

print("swap ETH -> DAI")
print("dx", dx)
print("dy", -dy)
print("x", x)
print("y", y)
print("spot2", spot2)
print("-------------")

fee = dy0 - dy1
print("loss", fee)
# Profit for borrowing DAI at spot1
eth_col = 1
print("profit", 0.8 * spot1 * eth_col - fee - spot2 * eth_col)
```
